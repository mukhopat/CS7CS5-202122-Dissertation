```python
from pydriller import Repository

repo_url = 'https://github.com/mukhopat/Automate_CI_PP1_Assessment_Form'

for commit in Repository(repo_url).traverse_commits():
    print(commit.hash)
    print(commit.msg)
    print(commit.author.name)

    for file in commit.modified_files:
        print(file.filename, ' has changed')
```

```python
from pydriller.metrics.process.contributors_experience import ContributorsExperience
metric = ContributorsExperience(path_to_repo=repo_url,
                                from_commit='257c4bedb7855a45aa4cd632d670d0702e9b03e0',
                                to_commit='a807be8c1038be66a1daaecd3eed41e8b87737b6')
files = metric.count()
print('Files: {}'.format(files))
```

```python
initial_commit = '257c4bedb7855a45aa4cd632d670d0702e9b03e0'
final_commit = 'a807be8c1038be66a1daaecd3eed41e8b87737b6'
```

```python
from pydriller.metrics.process.hunks_count import HunksCount
metric = HunksCount(path_to_repo=repo_url,
                    from_commit=initial_commit,
                    to_commit=final_commit)
files = metric.count()
print('Files: {}'.format(files))
```

```python
for commit in Repository(repo_url).traverse_commits():
    for file in commit.modified_files:
        print('Author {} modified {} in commit {}'.format(commit.author.name, file.filename, commit.hash))
```
