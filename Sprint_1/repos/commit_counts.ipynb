```python
from pydriller import Repository
from time import time
from datetime import timedelta
import os
from get_repo_name import get_repo_name

top_repos_path = 'top_repos/'
clone_path = '/clone/'

start_time = time()

for category in os.listdir(top_repos_path):
     
    category_commit_count = 0

    print ( f'Category: {category}:\n' )

    for repo in os.listdir(top_repos_path + '/' + category):

        if ( repo != '.ipynb_checkpoints' ):

            repo_name = get_repo_name ( os.path.basename ( repo ) )
            clone_path = f'/clone/{ repo_name }'
            
            
            
            
            
            path = top_repos_path + category + '/' + repo + clone_path

            count = 0
            for commit in Repository(path).traverse_commits():
                count += 1

            category_commit_count += count   
            print ( f'{repo_name}: {count}' )

    print ( f'{category} commit count: {category_commit_count}' )
        
    print ( '\n' )
    
print( f'Execution time = { str ( timedelta ( seconds = time() - start_time ) ) }' )
```
