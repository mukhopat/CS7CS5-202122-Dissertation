```python
import pandas as pd

dataset_path = 'datasets/mininet_grouped_by_author_date.csv'
init_df = pd.read_csv (dataset_path)
init_df.head()
```

```python
init_df.insert ( 1, 'year_month', pd.to_datetime(init_df['date']).dt.to_period('M') )
init_df.head()
```

```python
df = init_df.groupby(['year_month', 'author_code']).sum().reset_index(level=[0,1])

df.head()
```

```python
month_list = df['year_month'].unique()
author_list = df['author_code'].unique()
count = 0

def new_zero_row ( author, month ):
    
    new_row = {
        'year_month': month,
        'author_code': author,
        'commit_count': 0,
        'additions': 0,
        'deletions': 0,
        'agg_loc': 0,
        'agg_complexity': 0,
        'agg_token_count': 0,
        'dmm_unit_size': 0,
        'dmm_unit_complexity': 0,
        'dmm_unit_interfacing': 0
    }
    
    return new_row

for author in author_list:
    for month in month_list:
        if df [ ( df.author_code == author ) & ( df.year_month == month )].shape[0] == 0:
            count += 1
            df = df.append( new_zero_row(author, month), ignore_index = True )
            
df.shape
```

```python
df.sort_values ( ['author_code', 'year_month'], axis = 0, ascending = True, inplace = True )
df.reset_index ( inplace = True, drop = True )
df
```
