```python
import pandas as pd

dataset_path = 'datasets/mininet_modfiles.csv'
df = pd.read_csv (dataset_path)
df.head()
```

```python
def generate_ordinal_encoding_map ( series ):
    enc_map = series.unique()
    enc_map = dict ( enumerate( enc_map.flatten(), 1 ) )
    enc_map = { val: key for key, val in enc_map.items() }
    enc_map['None'] = 0
    return enc_map
```

```python
enc_map = generate_ordinal_encoding_map ( df["commit_author_email"] )
df["commit_author_enc"] = df["commit_author_email"].replace(enc_map)
df
```

```python
df[df['commit_author'] == 0]
```

```python
import numpy as np

df.drop ( df[df.commit_author_enc == 0].index, inplace=True )
df.dropna ( subset = ['file_loc', 'file_complexity', 'file_token_count'], inplace = True )
df.reset_index(drop = True, inplace = True)
df
```

```python
from datetime import datetime

date_list = []

for date_time_str in df['commit_datetime']:
    date_time_str = date_time_str[0:-6].replace('-', '/')
    date_time_obj = datetime.strptime(date_time_str, '%Y/%m/%d %H:%M:%S')
    date_list.append ( date_time_obj.date() )
    
df['commit_date'] = date_list
df.head()
```

```python
grouped_df = df.groupby(['commit_date', 'commit_author_enc']).agg({
    'commit_hash': 'nunique', 
    'file_lines_added': 'sum', 
    'file_lines_deleted': 'sum',
    'file_loc': 'sum',
    'file_complexity': 'sum',
    'file_token_count': 'sum',
    'dmm_unit_size': 'median',
    'dmm_unit_complexity': 'median',
    'dmm_unit_interfacing': 'median'
}
).reset_index(level=[0,1])

grouped_df.rename(columns = {
    'commit_date':'date', 
    'commit_author_enc':'author_code',
    'commit_hash': 'commit_count',
    'file_lines_added': 'additions',
    'file_lines_deleted': 'deletions',
    'file_loc': 'agg_loc',
    'file_complexity': 'agg_complexity',
    'file_token_count': 'agg_token_count'
}, inplace = True)

grouped_df
```

```python
import matplotlib.pyplot as plt
%matplotlib qt

fig, ax = plt.subplots(2, 3)

author_dict = {}
for author_code in grouped_df['author_code'].unique():    
    
    date = grouped_df['date'].loc[grouped_df['author_code'] == author_code].tolist()
            
    commit_count = grouped_df['commit_count'].loc[grouped_df['author_code'] == author_code].tolist()
    additions = grouped_df['additions'].loc[grouped_df['author_code'] == author_code].tolist()
    deletions = grouped_df['deletions'].loc[grouped_df['author_code'] == author_code].tolist()
    agg_loc = grouped_df['agg_loc'].loc[grouped_df['author_code'] == author_code].tolist()
    agg_complexity = grouped_df['agg_complexity'].loc[grouped_df['author_code'] == author_code].tolist()
    agg_token_count = grouped_df['agg_token_count'].loc[grouped_df['author_code'] == author_code].tolist()
    
    ax[0, 0].plot(date, commit_count, label = f"Author #{author_code}")
    ax[0, 1].plot(date, additions, label = f"Author #{author_code}")
    ax[0, 2].plot(date, deletions, label = f"Author #{author_code}")
    ax[1, 0].plot(date, agg_loc, label = f"Author #{author_code}")
    ax[1, 1].plot(date, agg_complexity, label = f"Author #{author_code}")
    ax[1, 2].plot(date, agg_token_count, label = f"Author #{author_code}")
    
fig.autofmt_xdate()

ax[0, 0].set_title("date vs commit count")
ax[0, 1].set_title("date vs additions")
ax[0, 2].set_title("date vs deletions")
ax[1, 0].set_title("date vs agg. LOC")
ax[1, 1].set_title("date vs agg. complexity")
ax[1, 2].set_title("date vs agg. token count")

ax[0, 0].set_xlabel("date")
ax[0, 1].set_xlabel("date")
ax[0, 2].set_xlabel("date")
ax[1, 0].set_xlabel("date")
ax[1, 1].set_xlabel("date")
ax[1, 2].set_xlabel("date")

ax[0, 0].set_ylabel("commit count")
ax[0, 1].set_ylabel("additions")
ax[0, 2].set_ylabel("deletions")
ax[1, 0].set_ylabel("agg. LOC")
ax[1, 1].set_ylabel("agg. complexity")
ax[1, 2].set_ylabel("agg. token count")

plt.show()
```
