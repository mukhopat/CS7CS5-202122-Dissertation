```python
import os
from get_repo_name import get_repo_name

top_repos_path = 'top_repos/'

repos = []

for dir in os.listdir(top_repos_path):
    path = top_repos_path + dir
    repos.append ( os.listdir(path) )
```

```python
for i in range ( len(repos) ):
    for j in range ( len(repos[0]) ):
        repos[i][j] = get_repo_name ( repos[i][j] )
```

```python
import pandas as pd

def transpose ( lst ):
    return list(map(list, zip(*lst)))

df = pd.DataFrame ( transpose(repos), columns = ['A', 'B'] )
df
```

```python
list(set(df.A) & set(df.B))
```
