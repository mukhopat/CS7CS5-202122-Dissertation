```python
import pandas as pd
import os
from get_repo_name import get_repo_name
from glob import glob
from shutil import copy

results_path = './results/'
cat_datasets = 'top_repos_by_star_count_datasets'
dst = results_path + cat_datasets + '/'

if cat_datasets not in os.listdir(results_path):
    os.mkdir ( dst )

cat_path = './top_repos/top_repos_by_star_count/'

for repo in os.listdir(cat_path):

    repo_name = get_repo_name ( os.path.basename ( repo ) )
    datasets_path = '/datasets/'
    csv_file = f'{ repo_name }_modfiles.csv'
    if csv_file in os.listdir ( cat_path + repo + datasets_path ):
        src = cat_path + repo + datasets_path + csv_file
        copy ( src, dst )
```

```python
glob_path = os.path.join ( dst, "*_modfiles.csv" )
master_dataset = glob ( glob_path )

df = pd.concat ( map ( pd.read_csv, master_dataset ), ignore_index = True )
df
```

```python
df.insert ( 1, 'year_month', pd.to_datetime(df['commit_datetime'], utc = True).dt.to_period('M') )
df
```

```python
from analysis_lib import year_pre_covid_ym, pre_covid_ym, covid_ym, year_post_covid_ym

df.drop ( df[df.year_month < year_pre_covid_ym].index, inplace = True )
df.drop ( df[df.year_month > year_post_covid_ym].index, inplace = True )
df.reset_index ( drop = True, inplace = True )
df
```

```python
from analysis_lib import generate_ordinal_encoding_map

enc_map = generate_ordinal_encoding_map ( df["commit_author_email"] )
df["commit_author_enc"] = df["commit_author_email"].replace(enc_map)
df.head()
```

```python
df [ df['commit_author_enc'] == 0 ].shape[0] # check if any invalid author code is present
```

```python
pre_df = df[df.year_month <= pre_covid_ym ]
pre_df.year_month.min(), pre_df.year_month.max()
```

```python
post_df = df[df.year_month >= covid_ym ]
post_df.year_month.min(), post_df.year_month.max()
```

```python
grouped_pre_df = pre_df.groupby(['commit_author_enc']).agg({
    
    'commit_hash': 'nunique', 
    'file_lines_added': 'sum', 
    'file_lines_deleted': 'sum',
    'file_loc': 'sum',
    'file_complexity': 'sum',
    'file_token_count': 'sum'
} ).reset_index(level=[0])


grouped_pre_df.rename(columns = {
    'commit_hash': 'commit_count',
    'file_lines_added': 'additions',
    'file_lines_deleted': 'deletions',
    'file_loc': 'agg_loc',
    'file_complexity': 'agg_complexity',
    'file_token_count': 'agg_token_count'
}, inplace = True)

grouped_pre_df
```

```python
grouped_post_df = post_df.groupby(['commit_author_enc']).agg({
    
    'commit_hash': 'nunique', 
    'file_lines_added': 'sum', 
    'file_lines_deleted': 'sum',
    'file_loc': 'sum',
    'file_complexity': 'sum',
    'file_token_count': 'sum'
} ).reset_index(level=[0])


grouped_post_df.rename(columns = {
    'commit_hash': 'commit_count',
    'file_lines_added': 'additions',
    'file_lines_deleted': 'deletions',
    'file_loc': 'agg_loc',
    'file_complexity': 'agg_complexity',
    'file_token_count': 'agg_token_count'
}, inplace = True)

grouped_post_df
```

```python
from analysis_lib import calc_and_compare_means

calc_and_compare_means ( grouped_pre_df, grouped_post_df )
```
