```python
import pandas as pd
import os
from get_repo_name import get_repo_name

repo_name = get_repo_name ( os.path.basename ( os.getcwd() ) )
dataset_path = f'datasets/{repo_name}_final.csv'
df = pd.read_csv (dataset_path)
df
```

```python
# df['year_month'] = pd.to_datetime(df['year_month'])
df = df.set_index(df['year_month'])
df
```

```python
pre_covid_ym = '2020-02'
covid_ym = '2020-03'
df1 = df [ str(df.year_month[0]) : pre_covid_ym ]
df2 = df [ covid_ym: str(df.year_month[-1]) ]
df1 = df1.reset_index ( drop = True )
df2 = df2.reset_index ( drop = True )

df1.shape, df2.shape
```

```python
from top_contributors import get_equilibrium_pc, get_pareto_pc

print("Pre-pandemic equilibrium percentages:")
for idx in range ( 1, len(df1.columns) ):
    print ( f'{df1.columns[idx]} = { get_equilibrium_pc ( list ( df1[df1.columns[idx]] ) ) }' )
```

```python
print("Post-pandemic equilibrium percentages:")
for idx in range ( 1, len(df2.columns) ):
    print ( f'{df2.columns[idx]} = { get_equilibrium_pc ( list ( df2[df2.columns[idx]] ) ) }' )
```

```python
print("Pre-pandemic Pareto percentages:")
for idx in range ( 1, len(df1.columns) ):
    print ( f'{df1.columns[idx]} = { get_pareto_pc ( list ( df1[df1.columns[idx]] ) ) }' )
```

```python
print("Post-pandemic Pareto percentages:")
for idx in range ( 1, len(df2.columns) ):
    print ( f'{df2.columns[idx]} = { get_pareto_pc ( list ( df2[df2.columns[idx]] ) ) }' )
```
